### Pydantic Models : Creating and Using Models

```python
from pydantic import BaseModel
```

```python
class Person(BaseModel):
    name: str
    age: int
    city: str

person = Person(name= "Monirul", age = 26, city = "Dhaka")
print(person)
```

### Model with Optional Value

```python
from typing import Optional

class Employee(BaseModel):
    name: str
    id: int
    salary: Optional[float] = None
    is_active: Optional[bool] = True

emp1 = Employee(name="Monirul",
                id = 13600,
                salary = 50000)

print(emp1)

```

```python
emp2 = Employee(name="Monirul",
                id = 13600,
                )

print(emp2)
```

### Model with List

```python
from typing import List

class Students(BaseModel):
    courseName : str
    studentList: List[str]
    capacity: Optional [int] = 25

students = Students(courseName= "Machine Learning",
                    studentList= ["Monirul", "Akib", "Shanto"])
print(students)
```
